# Image Crop
zoom $IN$ img part in zoom $OUT$ img .. using $alexnet-fc7$ feature 

```python
num_img_vis = 6
plt.rcParams['figure.figsize'] = (20,20)
root_path = '/media/lerrel/0A60802760801B97/Drone_Data/drone_script/training_progress/filtered_data/pred_vis/crop_vis/'
content = subprocess.check_output('ls ' + root_path, shell = True)
content = content.split('\n')
indx = random.sample(range(len(content) - 1), num_img_vis)

fig = plt.figure()
j = 1
for i in indx:
    plt.subplot(num_img_vis/2,2,j)
    j += 1
    im = cv2.imread(root_path + content[i])
    plt.imshow(cv2.cvtColor(im,cv2.COLOR_BGR2RGB))
    
plt.show()
```
